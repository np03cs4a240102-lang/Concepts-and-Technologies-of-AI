Task-1

```python
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

low = []
moderate = []
high = []

for record in time_data:
    study = record[0]
    if study < 3:
        low.append(study)
    elif 3 <= study <= 5:
        moderate.append(study)
    else:
        high.append(study)

print("Low study hours:", low)
print("Moderate study hours:", moderate)
print("High study hours:", high)
```

Task 4

```python
study_hours = []
entertainment_hours = []
sleep_hours = []

for study, ent, sleep in time_data:
    study_hours.append(study)
    entertainment_hours.append(ent)
    sleep_hours.append(sleep)

def avg(lst):
    return sum(lst) / len(lst) if lst else 0

avg_study = avg(study_hours)
avg_ent = avg(entertainment_hours)
avg_sleep = avg(sleep_hours)

print(f"Average study hours: {avg_study:.3f}")
print(f"Average entertainment hours: {avg_ent:.3f}")
print(f"Average sleep hours: {avg_sleep:.3f}")
```

```python
def sum_nested_list(nested_list):
    """
    Recursively sum all integers in a nested list structure.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            # assume element is numeric
            total += element
    return total

# Test
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Sum of nested_list:", sum_nested_list(nested_list))  # expected 36

# additional tests
print("Sum [1,[2,3],[4,[5]]]:", sum_nested_list([1,[2,3],[4,[5]]]))  # expected 15
```

```python
def generate_permutations(s):
    """
    Return a list of unique permutations of string s.
    Uses backtracking and handles duplicates.
    """
    results = []
    chars = sorted(s)  # sort to group duplicates
    used = [False] * len(chars)

    def backtrack(path):
        if len(path) == len(chars):
            results.append(''.join(path))
            return
        for i in range(len(chars)):
            if used[i]:
                continue
            # skip duplicates: if this char is same as previous and previous is not used in this branch
            if i > 0 and chars[i] == chars[i-1] and not used[i-1]:
                continue
            used[i] = True
            path.append(chars[i])
            backtrack(path)
            path.pop()
            used[i] = False

    backtrack([])
    return results

# Tests
print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```
